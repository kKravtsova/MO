# Вступ

---


При використанні прямих методів пошуку потрібно тільки
обчислення значень функції, але кількість цих обчислень може бути дуже
велике. Методи першого порядку використовують інформацію про
значеннях похідних функцій, тобто дають можливість знаходження
стаціонарних точок, які відповідають необхідній умові першого порядка.
Методи, що використовують похідні, носять ітеративний характер, так як
компоненти градієнта є нелінійними функціями. Градієнт вказує напрямки
зростання цільової функції. Оскільки методи використовують значення
похідних, передбачається, що f(x), ∇f(x), ∇^2f(x) – існують та неперервні.
Метод спряженого градієнта Флетчера-Рівса – будується
послідовність напрямків пошуку S, що є лінійними комбінаціями.

# 2. Теоритична частина

---


У методі сполученого градієнта Флетчера-Ривса будується
послідовність напрямків пошуку S, що є лінійними комбінаціями
1. −∇f(x), поточного напряму найшвидшого спуску;
2. S(0), S(1), …, S(n−1) попередніх напрямків пошуку.  
image.png

Вагові коефіцієнти γ(i) вибіраются таким чином, щоб напрямок S(k)
був Н-спряжений з усіма побудованими раніше напрямками пошуку. Тоді,
враховуючи спряженість напрямків, виходить, що для обчислення нового
напрямку пошуку в точці x(k) використовуються значення тільки поточного
та попереднього градієнтів. Всі вагові множники, що передують γ(k), а саме

```python
def norm(x1, x2):
    return sqrt(x1 ** 2 + x2 ** 2)
```

```python
func_counter = 0
def func(x1, x2):
    global func_counter
    func_counter += 1
    return np.sum(100.0*(x2-x1**2.0)**2.0 + (1-x1)**2.0)
```

```python
def sven(f, lamb, x, s, case, parameter):
    la = [lamb]
    #s_norm = norm(s[0], s[1])
    x_norm = norm(x[0], x[1])
    # d_lambd = 0.1 * (x_norm / s_norm)
    d_lambd = parameter * (x_norm / 1)
    # 0.1 -- исправить
    # визначення знаку перед дельта лямбда
    f_0 = f(x[0], x[1])
    f_r = f(x[0] + (lamb + d_lambd) * s[0], x[1] + (lamb + d_lambd) * s[1])
    f_l = f(x[0] + (lamb - d_lambd) * s[0], x[1] + (lamb - d_lambd) * s[1])
    y = [f_0]
    if f_l < f_0 < f_r:
        k = -1
        y.append(f_l)
        la.append(lamb - d_lambd)
    elif f_l > f_0 > f_r:
        k = 1
        y.append(f_r)
        la.append(lamb + d_lambd)
    elif f_l > f_0 and f_r > f_0:
        return [lamb - d_lambd, lamb, lamb + d_lambd]
    i = 1
    # знаходження інтервалу невизначеності
    while y[i] < y[i - 1]:
        la_i = la[i] + k * d_lambd * (2 ** i)
        la.append(la_i)
        y.append(f(x[0] + la_i * s[0], x[1] + la_i * s[1]))
        i += 1
    last4 = [la[i], (la[i] + la[i - 1]) / 2, la[i - 1], la[i - 2]]
    last4_evaluated = []
    for lambd in last4:
        last4_evaluated.append(f(x[0] + (lambd * s[0]), x[1] + (lambd * s[1])))
    yc = min(last4_evaluated)
    ind = last4_evaluated.index(yc)
    if case == 2:
        return [last4[ind - 1], last4[ind + 1]]
    else:
        if ind == 1:
            last3 = [last4[0], last4[1], last4[2]]
        if ind == 2:
            last3 = [last4[1], last4[2], last4[3]]
        return last3
```

```python
def dsk_powell(f, la, x, s, eps=0.01):
    # знаходження вершини нової параболи за алгоритмом ДСК
    la = sorted(la)
    #s_norm = norm(s[0], s[1])
    if len(la) == 3:
        x1 = la[0]
        x2 = la[1]
        x3 = la[2]
    else:
        x1 = la[0]
        x2 = sum(la) / 2
        x3 = la[1]
    dx = abs(x3 - x2)
    f1 = f(x[0] + x1 * s[0], x[1] + x1 * s[1])
    f2 = f(x[0] + x2 * s[0], x[1] + x2 * s[1])
    f3 = f(x[0] + x3 * s[0], x[1] + x3 * s[1])

    x_dsk = x2 + (dx * (f1 - f3)) / (2 * (f1 - 2 * f2 + f3))
    f_dsk = func(x[0] + x_dsk * s[0], x[1] + x_dsk * s[1])
    if abs(min([x1, x2, x3]) - x_dsk) < eps and abs(min([f1, f2, f3]) - f_dsk) < eps:
        return x_dsk
    else:
        while abs(min([x1, x2, x3, x_dsk]) - x_dsk) > eps and abs(min([f1, f2, f3, f_dsk]) - f_dsk) > eps:
            lis = sorted([x1, x2, x3, x_dsk])
            #print(lis)
            ind = lis.index(min(lis))
            if ind == 0:
                lis = [lis[ind], lis[ind + 1], lis[ind + 2]]
            elif ind == 3:
                lis = [lis[ind - 2], lis[ind - 1], lis[ind]]
            else:
                lis = [lis[ind - 1], lis[ind], lis[ind + 1]]
            x1 = lis[0]
            x2 = lis[1]
            x3 = lis[2]
            #print(x1, x2, x3)
            f1 = f(x[0] + x1 * s[0], x[1] + x1 * s[1])
            f2 = f(x[0] + x2 * s[0], x[1] + x2 * s[1])
            f3 = f(x[0] + x3 * s[0], x[1] + x3 * s[1])
            a1 = (f2 - f1) / (x2 - x1)
            a2 = (1 / (x3 - x2)) * ((f3 - f1) / (x3 - x1) - (f2 - f1) / (x2 - x1))
            x_dsk = (x1 + x2) / 2 - a1 / (2 * a2)
            #print(x_dsk)
            f_dsk = f(x[0] + x_dsk * s[0], x[1] + x_dsk * s[1])
            if abs(min([x1, x2, x3, x_dsk]) - x_dsk) < eps and abs(min([f1, f2, f3, f_dsk]) - f_dsk) < eps:
                break
            """f_list = [f1, f2, f3, f_dsk]
            list_x = [x1, x2, x3, x_dsk]
            f_min = min(f_list)
            inx_min = f_list.index(f_min)
            x_min = list_x[inx_min]
            if abs(f_min - f_dsk) < eps and abs(x_min - x_dsk) < eps:
                break"""
        return x_dsk
```

```python
def gold(f, x, la, s, eps=0.01):
    la = sorted(la)
    # s_norm = norm(s[0], s[1])
    if len(la) == 3:
        a = la[0]
        b = la[2]
    else:
        a = la[0]
        b = la[1]
    x1 = a + 0.382 * (b - a)
    x2 = a + 0.618 * (b - a)
    while abs(x1 - x2) > eps:
        f1 = f(x[0] + x1 * s[0], x[1] + x1 * s[1])
        f2 = f(x[0] + x2 * s[0], x[1] + x2 * s[1])
        if f1 < f2:
            b = x2
        else:
            a = x1
        x1 = a + 0.382 * (b - a)
        x2 = a + 0.618 * (b - a)

    return (a + b) / 2
```

```python
def grad(f, x1, x2, h, typ):
    if typ == "c":
        diff_x = (f(x1 + h, x2) - f(x1 - h, x2)) / 2 / h
        diff_y = (f(x1, x2 + h) - f(x1, x2 - h)) / 2 / h

    if typ == "l":
        diff_x = (f(x1, x2) - f(x1 - h, x2)) / h
        diff_y = (f(x1, x2) - f(x1, x2 - h)) / h

    if typ == "r":
        diff_x = (f(x1 + h, x2) - f(x1, x2)) / h
        diff_y = (f(x1, x2 + h) - f(x1, x2)) / h

    return [diff_x, diff_y]
```

```python
def S1_search(grad_f0, grad_f1, s0):
    anti = [-1*i for i in grad_f1]
    return add(anti, dot(dot(transpose(grad_f1), grad_f1), s0) / dot(transpose(grad_f0), grad_f0))
```

```python
def gamma(grad_k, grad_k_1):
    return dot(transpose(subtract(grad_k, grad_k_1)), grad_k) / norm(grad_k_1[0], grad_k_1[1])
```

```python
def fletcher_ryvs(f, x, s, h, typ, choice, sven_parameter, eps, ending, counter=1):
    print("====================== КРОК %d ======================" % counter)
    print("Пошук x%d" % counter)
    print("S%d =" % (counter - 1), s)
    la = sven(f, 0, x, s, choice, sven_parameter)
    print("Інтервал невизначеності: ", la)
    if choice == 1:
        lmbd = dsk_powell(f, la, x, s, eps)
        if lmbd <= 0:
            anti_grad = [-1*i for i in s]
            s = anti_grad
        print("Оптимальне значення кроку методом ДСК-Пауелла: ", lmbd)
    if choice == 2:
        lmbd = gold(f, x, la, s, eps)
        if lmbd <= 0:
            anti_grad = [-1*i for i in s]
            s = anti_grad
        print("Оптимальне значення кроку методом Золотого перетину: ", lmbd)
    x1 = [x[0] + lmbd * s[0], x[1] + lmbd * s[1]]
    print("Координати x%d =" % counter, x1)
    grad_f1 = grad(f, x1[0], x1[1], h, typ)
    print("Градієнт на кроці %d =" % counter, grad_f1)
    grad_f0 = [-1*i for i in s]
    s1 = S1_search(grad_f0, grad_f1, s)
    print("Вектор S%d =" % counter, s1)
    if ending == 1:
        criteria1 = norm(x1[1] - x[1], x1[0] - x[0]) / norm(x[0], x[1])
        criteria2 = abs(f(x1[0], x1[1]) - f(x[0], x[1]))
        if criteria1 < eps and criteria2 < eps:
            print("\nУмова закінчення виконується!")
            print("Значення функції в точці x%d =" % counter, x1)
            print("f = ", f(x1[0], x1[1]))
        elif lmbd == 0:
            print("\nКрок лямбда дорівнює 0!")
            print("Значення функції в точці x%d =" % counter, x1)
            print("f = ", f(x1[0], x1[1]))
        else:
            try:
                return fletcher_ryvs(func, x1, s1, h, typ, choice, sven_parameter, eps, ending, counter + 1)
            except RuntimeError:
                print("\nВелика кількість рекурсії! Умова закінчення не виконується!")
                print("Значення функції в точці x%d =" % counter, x1)
                print("f = ", f(x1[0], x1[1]))
    if ending == 2:
        criteria = norm(s1[0], s1[1])
        if criteria < eps:
            print("\nУмова закінчення виконується!")
            print("Значення функції в точці x%d =" % counter, x1)
            print("f = ", f(x1[0], x1[1]))
        elif lmbd == 0:
            print("\nКрок лямбда дорівнює 0!")
            print("Значення функції в точці x%d =" % counter, x1)
            print("f = ", f(x1[0], x1[1]))
        else:
            try:
                return fletcher_ryvs(func, x1, s1, h, typ, choice, sven_parameter, eps, ending, counter + 1)
            except RuntimeError:
                print("\nВелика кількість рекурсії! Умова закінчення не виконується!")
                print("Значення функції в точці x%d =" % counter, x1)
                print("f = ", f(x1[0], x1[1]))

    return x1
```

```python
def g_ext(x1,x2):
    return 4 - (x1 - 2)**2 - x2**2

def P_ext(x1, x2):
    return func(x1, x2) + R * g_ext(x1,x2) ** 2

def P_int(x1, x2):
    return func(x1, x2) + R * (4 - (x1 + 2) ** 2 - x2 ** 2) ** 2
```
